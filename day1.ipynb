```python
score = float(input("Enter the score"))
print(f"Score is: {score}")

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
elif score >= 60:
    print("Grade: D")
else:
    print("Grade: F")
```

```python
has_ticket = True
has_id = True
print(f"Has Ticket: {has_ticket}, Has ID: {has_id}")

if has_ticket and has_id:
    print("Entry allowed.")
elif has_ticket or has_id:
    print("Partial requirements met. Check with staff.")
    if not has_id:
        print("Warning: ID is missing!")
   
else:
    print("Entry denied.")
```
